1. Memuat Data

```python
import pandas as pd

# Memuat data dari file CSV
df = pd.read_csv('movie_sample_dataset.csv')
```

2. Memeriksa Data

```python
# Menampilkan 5 baris pertama dari dataset
df.head()
```

```python
# Memeriksa informasi umum dataset
df.info()
```

```python
# Identifikasi nilai yang hilang
import numpy as np

# Mengganti "?" dengan NaN
df.replace("?", np.nan, inplace=True)

# Memeriksa jumlah missing values di setiap kolom
print(df.isnull().sum())
```

3. Membersihkan Data

```python
# Menghapus baris dengan missing values pada kolom "gross"
df.dropna(subset=["gross"], axis=0, inplace=True)

# Menghapus baris dengan missing values pada kolom "budget"
df.dropna(subset=["budget"], axis=0, inplace=True)

# Menghapus baris dengan missing values pada kolom "genres"
df.dropna(subset=["genres"], axis=0, inplace=True)

# Menghapus baris dengan missing values pada kolom "director_name"
df.dropna(subset=["director_name"], axis=0, inplace=True)

# Reset index setelah menghapus baris
df.reset_index(drop=True, inplace=True)

# Melihat jumlah missing values di setiap kolom
print(df.isnull().sum())
```

```python
# Melihat nilai yang paling sering muncul pada kolom color
modus_color = df["color"].mode()
print(modus_color[0])
```

```python
# Mengganti missing values pada kolom color dengan nilai modus
df["color"].replace(np.nan, "Color", inplace=True)

# Melihat jumlah missing values di setiap kolom
print(df.isnull().sum())
```

```python
# Menyeragamkan pola huruf pada kolom color
df["color"] = df["color"].str.title()

# Menampilkan 5 baris pertama dari dataset
df.head()
```

```python
# Memeriksa informasi umum dataset
df.info()
```

```python
# Mencari letak nilai yang tidak standar (negatif)
for col in df.select_dtypes(include=["int", "float"]).columns:
    print(f"\nKolom {col}:")
    print(df[df[col] < 0])
```

```python
# Mengubah nilai negatif jadi nilai positif
df.loc[5, "duration"] = abs(df.loc[5, "duration"])
df.loc[8, "imdb_score"] = abs(df.loc[8, "imdb_score"])
df.loc[67, "imdb_score"] = abs(df.loc[67, "imdb_score"])

# Cek letak nilai negatif lagi
for col in df.select_dtypes(include=["int", "float"]).columns:
    print(f"\nKolom {col}:")
    print(df[df[col] < 0])
```

Sudah tidak ditemukan nilai negatif

4. Transformasi

```python
# Konversi tipe data kolom budget dan gross ke tipe data int
df[["budget", "gross"]] = df[["budget", "gross"]].astype(int)

# Menampilkan tipe data per kolom
df.dtypes
```

```python
# Memisahkan kolom genre
max_genres = df["genres"].str.split("|").str.len().max()

for i in range(max_genres):
    df[f"genres_{i+1}"] = df["genres"].str.split("|").str[i].fillna("-")

# Menghapus kolom genre asli (yang belum dipisahkan)
df.drop("genres", axis=1, inplace=True)

# Menampilkan 5 baris pertama dataset
df.head()
```

```python
# Mengubah pola huruf menjadi huruf kapital total pada data bertipe string pada kolom tertentu
for col in df.columns:
    if df[col].dtype == 'object':  # Cek jika kolom bertipe string
        df[col] = df[col].str.upper()

# Menampilkan 5 baris pertama dataset
df.head()
```

```python
# Normalisasi data kolom gross dan budget dengan mengubah skala ke rentang 0-1
df["gross"] = df["gross"] / df["gross"].max()
df["budget"] = df["budget"] / df["budget"].max()

# Menampilkan 5 baris pertama dataset
df.head()
```

```python
# Menghapus data duplikat berdasarkan judul yang sama
df = df.drop_duplicates(subset=["movie_title"])

# Reset index setelah menghapus baris
df.reset_index(drop=True, inplace=True)

# Memeriksa informasi umum dataset
df.info()
```
